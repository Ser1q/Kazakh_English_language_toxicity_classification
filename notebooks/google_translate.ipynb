```python
df = pd.read_csv('../data/processed/combined_balanced.csv').sample(n=1688, random_state=42).reset_index(drop=True)

df.info()
```

```python
def count_chars(text):
    return len(text)
```

```python
texts = df['clean_text'][2688:].fillna("").astype(str)
count_char = texts.apply(count_chars)
print(count_char.sum())
```

```python
def translate(text):
    if isinstance(text, bytes):
        text = text.decode("utf-8")
    
    result = translate_client.translate(text, target_language='kk')
    return result['translatedText']
```

```python
df['kk_clean_text'] = texts.apply(translate)
```

```python
import re 
def is_mostly_english(text):
    return len(re.findall(r'[a-zA-Z]', text)) / (len(text) + 1e-5) > 0.5

def is_garbled(text):
    return bool(re.search(r'(.)\1{4,}', text))  # e.g. "loooool"

def is_too_short(text):
    return len(text.split()) < 3

def needs_manual_check(text):
    return is_mostly_english(text) or is_garbled(text) or is_too_short(text)

# Example
df['needs_check'] = df['kk_clean_text'].astype(str).apply(needs_manual_check)

# Now you can view and inspect rows needing manual verification
problematic_rows = df[df['needs_check'] == True]
```

```python
print(len(problematic_rows['kk_clean_text']))
```

```python
df = df.drop(df[df['needs_check'] == True].index)
```
